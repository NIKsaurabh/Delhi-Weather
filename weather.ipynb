# Creating a dataset which only contains date and minimum temperature

### Removing time from 'datetime_utc' and convering the dataset in different csv file as 'date_temperature.csv'.
### I have converted to different csv file because it takes time to ceonvert each data of a column to different format.
date_temperature=dataset[['datetime_utc','minimum temperature']]

for i in date_temp['datetime_utc']:
    
    t=i[:10]
    #print(t)
    date_temp['datetime_utc'] = date_temp['datetime_utc'].replace(i,t)

date_temperature = round(date_temp.groupby('datetime_utc').mean())

date_temperature.to_csv('date_temperature.csv')

```python
date_temperature = pd.read_csv('date_temperature.csv',index_col=0)
```

```python
plt.figure(figsize=(20,10))
date_temperature.plot()
plt.xticks(rotation='vertical')
plt.show()
```

```python
#calculating rolling mean with 12 windows 
roll_mean = date_temperature['minimum temperature'].rolling(window = 12).mean()
```

```python
#calculating rolling standard deviation with 12 windows 
roll_std = date_temperature['minimum temperature'].rolling(window = 12).std()
```

#### Plotting rolling statistics

```python
plt.figure(figsize=(20,10))
#plotting original data
plt.plot(date_temperature,color = 'blue', label='Original')
plt.plot(roll_mean,color='red',label='Rolling mean')
plt.plot(roll_std,color='black',label='rolling standard deviation')
plt.xticks(rotation='vertical')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.title('Original dataset vs rolling mean vs rolling standard deviation')
plt.legend(loc='best')
plt.show(block=False)
```

```python
roll_std
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(date_temperature['minimum temperature'], autolag='AIC')
print('ADF Statistic: '+str(result[0]))
print('n_lags: '+str(result[1]))
print('p-value: '+str(result[1]))
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')   
```

Hence difference between ADF Statistics and critical values are very high, so datadrame is not stationary

```python
interval=1
diff=[]
for i in range(interval,len(date_temperature)):
    value=date_temperature['minimum temperature'][i]-date_temperature['minimum temperature'][i-interval]
    diff.append(value)
diff
```
